```python
import pandas as pd

# File paths
datasets = {
    "Facebook Posts": "Downloads/period_03/2024_fb_posts_president_scored_anon.csv",
    "Twitter Posts": "Downloads/period_03/2024_tw_posts_president_scored_anon.csv"
}

# Function to analyze with Pandas
def analyze_with_pandas(path, name):
    print(f"\n\n📊 === Analyzing {name} ===")
    df = pd.read_csv(path)

    print("\n=== Descriptive Statistics ===")
    print(df.describe(include='all'))

    print("\n=== Unique Values Per Column ===")
    print(df.nunique())

    print("\n=== Most Common Values ===")
    for col in df.select_dtypes(include='object').columns:
        print(f"\n{col}:\n", df[col].value_counts().head(3))

# Run for all three datasets
for name, path in datasets.items():
    analyze_with_pandas(path, name)
```

## Analyzing data between Datasets

```python
import pandas as pd

# Shared columns
shared = [
    'advocacy_msg_type_illuminating', 'attack_msg_type_illuminating',
    'covid_topic_illuminating', 'cta_msg_type_illuminating',
    'economy_topic_illuminating', 'education_topic_illuminating',
    'engagement_cta_subtype_illuminating', 'environment_topic_illuminating',
    'foreign_policy_topic_illuminating', 'fraud_illuminating',
    'freefair_illuminating', 'fundraising_cta_subtype_illuminating',
    'governance_topic_illuminating', 'health_topic_illuminating',
    'image_msg_type_illuminating', 'immigration_topic_illuminating',
    'incivility_illuminating', 'issue_msg_type_illuminating',
    'lgbtq_issues_topic_illuminating', 'military_topic_illuminating',
    'race_and_ethnicity_topic_illuminating', 'safety_topic_illuminating',
    'scam_illuminating', 'social_and_cultural_topic_illuminating',
    'technology_and_privacy_topic_illuminating',
    'voting_cta_subtype_illuminating', 'womens_issue_topic_illuminating'
]

# Paths for your datasets
datasets = {
    "Facebook Ads": "Downloads/period_03/2024_fb_ads_president_scored_anon.csv",
    "Facebook Posts": "Downloads/period_03/2024_fb_posts_president_scored_anon.csv",
    "Twitter Posts": "Downloads/period_03/2024_tw_posts_president_scored_anon.csv"
}

def analyze_shared(df, name):
    print(f"\n📊 Shared-column analysis for: {name}")
    sub = df[shared]
    # Numeric statistics (mean, median, min, max, std)
    print("\n🔹 Numeric summary:")
    print(sub.describe().T[['mean', '50%', 'min', 'max', 'std']])
    # Unique counts
    print("\n🔹 Unique values per column:")
    print(sub.nunique())
    # Top 3 most common values
    print("\n🔹 Top 3 most common for each shared column:")
    for col in shared:
        top = sub[col].value_counts().head(3)
        print(f"• {col}:\n{top}\n")

# Run for each dataset
for name, path in datasets.items():
    df = pd.read_csv(path)
    analyze_shared(df, name)
```
